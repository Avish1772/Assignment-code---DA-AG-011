Question 1 : What is Dimensionality Reduction? Why is it important in machine
learning?
Dimensionality Reduction is the process of reducing the number of input features (variables) in a dataset while preserving as much useful information as possible.

In simple words:
We shrink the dataset from high dimensions to fewer dimensions without losing important patterns.
1. Reduces Overfitting

When there are too many features, the model may memorize noise instead of learning patterns.
Dimensionality reduction removes irrelevant/noisy features → improves generalization.

2. Improves Model Performance

Less data → faster training → less computation.
Models become simpler and more efficient.

3. Handles the “Curse of Dimensionality”

High-dimensional data becomes sparse and distances lose meaning.
Reducing dimensions helps algorithms like KNN, K-means, etc., perform better.

4. Better Visualization

We cannot visualize data with 50, 100, or 1000 features.
Dimensionality reduction (e.g., using PCA or t-SNE) helps convert data into 2D or 3D, which is easy to explore and understand.

5. Removes Multicollinearity

When two or more features are highly correlated, they carry duplicate information.
Dimensionality reduction combines them into fewer uncorrelated components

Question 2: Name and briefly describe three common dimensionality reduction
techniques..
1. Principal Component Analysis (PCA)

PCA is a linear dimensionality reduction method.

It transforms the data into new variables called principal components.

These components capture the maximum variance in the data.

The first few components keep most of the important information.

2. t-SNE (t-Distributed Stochastic Neighbor Embedding)

t-SNE is a non-linear technique mainly used for visualizing high-dimensional data in 2D or 3D.

It preserves local structure (i.e., points that are close remain close).

Great for visualizing clusters in datasets like images, text embeddings, etc.

3. LDA (Linear Discriminant Analysis)

LDA is a supervised dimensionality reduction technique.

It focuses on maximizing class separation.

It projects data into a lower-dimensional space where different classes become more distinguishable.

Often used in classification tasks like face recognition

Question 3: What is clustering in unsupervised learning? Mention three popular
clustering algorithms.

Clustering is an unsupervised learning technique used to group similar data points together based on patterns or features—without using labels.

In simple terms:

Clustering automatically divides data into groups (clusters) where items in the same group are more similar to each other than to items in other groups.

It is used in:

Customer segmentation

Image grouping

Anomaly detection

Recommender systems

Three Popular Clustering Algorithms
1. K-Means Clustering

Partitions data into K clusters.

Uses distance (usually Euclidean) to assign points to the nearest cluster center.

Fast and widely used.

2. Hierarchical Clustering

Builds a tree-like structure (dendrogram) of clusters.

Two types: Agglomerative (bottom-up) and Divisive (top-down).

No need to specify number of clusters initially.

3. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

Groups points based on density.

Can detect arbitrary-shaped clusters.

Can identify noise/outliers, unlike K-Mean

Question 5: List and briefly describe three types of anomaly detection techniques
1. Statistical Methods

These methods assume the data follows a probability distribution (e.g., normal distribution).

Points that lie far from the mean (e.g., more than 3 standard deviations) are considered anomalies.

Example: Z-score, Gaussian models.

2. Distance-Based Methods

These methods measure how far a data point is from its neighbors.

If a point is too far from others, it is labeled as an outlier.

Works well when anomalies are isolated.

Example: KNN-based anomaly detection, Mahalanobis distance.

3. Density-Based Methods

These methods check how densely populated a region is.

Points in low-density areas are considered anomalies.

Useful for complex, non-linear datasets.

Example: DBSCAN, Local Outlier Factor (LOF).

Question 6: What is time series analysis? Mention two key components of time series
data

What is Time Series Analysis?

Time series analysis is a method used to study data collected over time to identify patterns, trends, and relationships.

In simple words:

It analyzes how a value changes over time to forecast future values.

Examples:

Stock prices

Weather data

Sales data

Temperature readings

Two Key Components of Time Series Data
1. Trend

The long-term upward or downward movement in the data.

Example: Sales increasing every year.

2. Seasonality

Regular, repeating patterns in data over fixed periods.

Example: Ice cream sales rising every summer.

Question 7: Describe the difference between seasonality and cyclic behavior in time
series

Seasonality refers to patterns in a time series that repeat at fixed, regular intervals, such as daily, monthly, or yearly. These patterns are predictable and occur due to consistent factors like weather changes, holidays, or cultural events. In contrast, cyclic behavior represents long-term fluctuations that do not follow a fixed or regular schedule. Cycles vary in length and are often influenced by broader economic or business factors, such as periods of recession or growth. Unlike seasonality, cyclic patterns are irregular and harder to predict because their duration and timing are not consistent

Question 8: Write Python code to perform K-means clustering on a sample dataset.
(Include your Python code and output in the code box below.)


```python
# Generate a sample dataset
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=200, centers=3, random_state=42)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)
centers = kmeans.cluster_centers_

# Print cluster centers and first few labels
centers, labels[:10]

# Plotting the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centers[:, 0], centers[:, 1], marker='x', s=200)
plt.title("K-Means Clustering Plot")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

Question 9: What is inheritance in OOP? Provide a simple example in Python.

Inheritance is an important concept in Object-Oriented Programming (OOP) where a new class (called child or derived class) can inherit properties and methods from an existing class (called parent or base class

```python
# Parent class
class Animal:
    def speak(self):
        return "Animal makes a sound"

# Child class inheriting from Animal
class Dog(Animal):
    def speak(self):
        return "Dog barks"

# Creating objects
a = Animal()
d = Dog()

print(a.speak())  # Output: Animal makes a sound
print(d.speak())  # Output: Dog barks
```

## **Q10: How can time series analysis be used for anomaly detection?**

Time series analysis can be used for anomaly detection by examining patterns, trends, and behaviors in data collected over time and identifying points that deviate significantly from normal patterns. By understanding the expected behavior—such as trend, seasonality, or historical averages—time series models can flag unusual observations that fall outside predicted ranges. Techniques like moving averages, ARIMA models, and forecasting-based methods can be applied to generate expected values; any time point with a large prediction error or sudden abnormal spike or drop may be treated as an anomaly. Additionally, time series decomposition helps isolate irregular components, making it easier to detect unexpected changes. This approach is widely used in areas like fraud detection, network monitoring, sensor fault detection, stock market analysis, and healthcare monitoring.
